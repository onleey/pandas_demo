## at
이름으로 한 위치의 값만 가지고 온다.

```python
print(data.at[0,'tip'])
print(data.loc[0,'tip'])
print(data.iloc[0,1])
print(data['tip'][0])

#print(data.at[0]['tip'])  # 안됨.
print(data.loc[0]['tip'])
print(data.iloc[0][1])
print('---------------------------------------------')
print(data['tip']) # 열 (이름)
print('---------------------------------------------')
print(data[3:5])   #행 (번호)
```

```python
#at 은 slicing을 사용 할 수 없다. #InvalidIndexError
data.at[0:3,'tip']
```

```python
#at 은 fancy indexing 을 사용 할 수 없다.
#InvalidIndexError
data.at[[0,1,5],'tip']
```

```python
#열일때
#data[['tip','day']] #ok
#data['tip':'day'] #타입에러

#행일때
data[0:5]  # ok
#data[[0,1,3]] #키에러
```
